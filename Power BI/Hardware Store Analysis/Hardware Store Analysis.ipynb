## Analysis

```python
# Creating PRODUCTION_COST column

df['PRODUCTION_COST'] = df['STANDARD_COST'] * df['QUANTITY']
df[['STANDARD_COST', 'QUANTITY', 'PRODUCTION_COST']].head(10)
```

```python
# Creating REVENUE column
df['REVENUE'] = df['LIST_PRICE'] * df['QUANTITY']
df[['LIST_PRICE', 'QUANTITY', 'REVENUE']].head(10)
```

### Production Costs by Country and State

```python
df.groupby(by = ['COUNTRY_NAME', 'STATE']).agg({'PRODUCTION_COST':'sum'})
```

### Revenue by Category

```python
df.groupby(by='CATEGORY_NAME')['REVENUE'].sum()
```

### Top 10 Product Revenue per Category

```python
df.groupby('PRODUCT_ID')['REVENUE'].sum(numeric_only=True).nlargest(10)
```

### Inventory by Category and Product

```python
df.groupby(['CATEGORY_NAME', 'PRODUCT_ID'])['QUANTITY'].sum(numeric_only=True)
```
